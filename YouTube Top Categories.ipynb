```python
# Viewing the data ('pretty print')

with open('DE_category_id.json') as json_file:
    Germany = json.load(json_file) 
    print(json.dumps(Germany, indent=4, sort_keys=True))
```

```python
#Extracting data

with open('DE_category_id.json') as json_file:
    Germany = json.load(json_file) 
#     print(json.dumps(US, indent=4))
#Iterating through the given json dump and extracting the Kind, Id and title values

for i in range(len(Germany.get('items'))):
    print('Kind'+'-----'+Germany.get('items')[i].get('kind'))
    print('Id'+'-----'+Germany.get('items')[i].get('id'))
    print('Title'+'-----'+Germany.get('items')[i].get('snippet')['title'])
    print()
```

```python
#Extracting Titles only

for i in range(len(Germany.get('items'))):
    print('Title'+'-----'+Germany.get('items')[i].get('snippet')['title'])
    print()
```
