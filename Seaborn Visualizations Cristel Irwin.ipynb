In this assignment we will be using popular Titanic disaster data set and produce various charts.  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Load the data set 
df_titanic = pd.read_csv("titanic.csv")
```

```python
df_titanic.head()
```

```python
df_titanic.columns
```

#### Q1

```python
# count plot of sex vs. class and survived. 
# your code here

sns.catplot(x ='Sex', hue ='Survived',
kind ='count', col ='Pclass', data = df_titanic)
```

#### Q2

```python
## Distribution of Age

plt.figure(figsize=(8, 3))
plt.hist(df_titanic["Age"], bins=20)
plt.title(" ")
```

#### Q3

```python
# your code here

median=dict(color='green')
boxprops=dict(color='blue')

data = []

for i in range(1,4):
    data.append(df_titanic[df_titanic['Pclass'] == i]['Age'].dropna().values)

    
plt.boxplot(data, medianprops=median, boxprops=boxprops);
plt.title("Age")
plt.suptitle("Boxplot grouped by Pclass")
plt.xlabel('Pclass', fontsize=10)
plt.grid()
```

#### Q4

```python
### Your code here

sns.displot(data = df_titanic , x = "Age" , hue = "Pclass", hue_order = ['Female', 'Male'])
```

#### Q5

```python
# Your code here

grid = sns.FacetGrid(df_titanic, col='Survived', row='Pclass', height=2.0, aspect=1.0)
grid.map(plt.hist, 'Age', alpha=1.0, bins=25)
grid.add_legend();
```

#### Q6

```python
sns.factorplot(x = "Sex", y = "Age", data = df_titanic, kind = "box")
plt.show()
```

```python
#Changed plot to catplt to get rid of warning

sns.catplot(x = "Sex", y = "Age", data = df_titanic, kind = "box")
plt.show()
```

#### Q7

#### Plot the distribution of three variables
Fare, Age and Sex

```python
## Your code here

fig = plt.gcf()
fig.set_size_inches(10, 5)

sns.scatterplot(data = df_titanic, x = 'Fare', y = 'Age', hue = 'Survived') 
plt.xlabel("Fare")
plt.ylabel("Age")
```

#### Q8

Plot the distribution of fare paid by passengers according to age, survival and sex.

```python
# your code here

fig = plt.gcf()
fig.set_size_inches(10, 5)

sns.scatterplot(data = df_titanic, x = 'Fare', y = 'Age', hue = 'Survived', size = "Sex", sizes = (20, 80)) 
plt.xlabel("Fare")
plt.ylabel("Age")
```
