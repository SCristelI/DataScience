#### Q8 Create the follwing using subplots. Adjust the spacing. Play with Labels and title 


```python
fig, ax = plt.subplots(2, 2)
fig.tight_layout(h_pad = 4) # Adjusts spacing

ax[0, 0].set_title('Upper Left', color ="#581845", loc='left') #Changing label locations and colors
ax[0, 1].set_title('Upper Right', color ="#C70039", loc='right')
ax[1, 0].set_title('Lower Left', color ="#FFC300", loc='left')
ax[1, 1].set_title('Lower Right', color ="#DAF7A6",loc='right')

fig.suptitle('Main Title', color ="#2094F4", style='italic', fontname="Times New Roman", fontsize=20) #Adding main title, adjusting spacing, changing color and font
plt.subplots_adjust(top=0.85)

plt.show()
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2

plt.figure(figsize=(4, 2))
plt.plot(x, y,color = 'r')

plt.title("Line Plot using np.linspace", color ="#11436C", fontsize=12)

plt.xlabel('x', color ="#11436C", fontsize=10 )
plt.ylabel('y', color ="#11436C", fontsize=10)
plt.show()
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2

plt.figure(figsize=(5, 2))
ax = plt.subplot() 
ax.bar(x, y, width=1, color ="#C70039")

plt.title('Bar Plot using np.linspace')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

#### Q9
Create the following graph using the following properties

We can configure the ranges of the axes using the set_ylim and set_xlim methods in the axis object, or axis('tight') for automatically getting "tightly fitted" axes ranges:

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))


axes[0].plot(x, x**2, label = "x**2")
axes[0].plot(x, x**3, label = "x**3")
axes[0].legend(loc=2);
axes[0].set_title("default axes ranges")


axes[1].plot(x, x**2, label = "x**2")
axes[1].plot(x, x**3, label = "x**3")
axes[1].legend(loc=2);
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, label = "x**2")
axes[2].plot(x, x**3, label = "x**3")
axes[2].legend(loc=1);
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");

fig.suptitle('multiple plots') 
plt.subplots_adjust(top=0.85)

plt.show()
```
