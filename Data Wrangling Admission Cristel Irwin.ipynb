The data sources were mostly found online by searching "mock college applications". I searched for applicants for several different types of programs from universities across the United States. I changed the names of the applicants as well as the universities. For some of the files, I added embellishments such as hobbies and GPA scores. For a few of the candidates who already had STEM skills, I changed the wording to more common terms that I found in other applicant files. 

In order to filter out the best candidates, I used a 'for loop' to find the keywords from the Admissions Criteria file. This is printed under the line "Skills to look for in an applicant". I then subsetted keywords from this statment and placed them in a variable called "others". 

In the second 'for loop', I wrote a code that tells the computer to match the keywords from "others" to the Applications (App). The code returns all applicants with atleast one or more skills mentioned in "others".I used this medthod because this code can be changed to find applicants with 3 or more, 4 or more, 5 or more, etc. skills as well. 

I specifcially looked 'r', 'programming', 'python', 'algebra', 'analytical', 'math', 'troubleshoot', 'errors', 'analytical', 'computer', 'communicate','java','c++'. These were common words that I also found in real-life applications for other data science programs. I also attempted to filter out applcants based on GPA score but only got errors. 

The results are five applicants who have one or more combinations of keywords in their applications. Each applicant has a near perfect GPA score as well as several of the required skills needed for the program. 

```python
import pandas as pd

# Open csv, skip first 3 rows
df = pd.read_csv("Candidates.csv",delimiter='\t') 
print("All applicants:\n", df)

#Code to get the required skills from Criteria text

to_find = "Required skills" #define search term

with open("AdmissionsCriteria.txt") as a_file:
    for line in a_file:
        if to_find in line:
                required_skills = line
                print("\nSkills to look for in an applicant:", required_skills*1, sep="\n") # *1 to print once

# Creating a container to hold list of recommendations
recommended_candidate = []

apps_subset = df['Applications']
Cands = df['Candidate']
others = ['r','programming','python','algebra','analytical','math','troubleshoot','errors','analytical','computer','communicate','java','c++']

#Set count to zero to count recommendations
count=0
#Loop to traverse through csv column 'Applications' and return applicants
for f in apps_subset:

    App = open("{}.txt".format(f), "r")
    lines = App.read()
    skills_of_applicant = lines.split(' ')

    number_of_found_skills = 0

    for skill in skills_of_applicant:
        if skill.strip() in required_skills and skill.strip().lower() in others:
            number_of_found_skills += 1 #applicant has atleast 1 skills

    if number_of_found_skills >=2: #applicant has atleast 2 required skills
        count += 1
        print("\nRecommended Candidate", count,":\n", f)
        print(lines, sep="\n")
        

```
